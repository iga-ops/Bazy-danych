# Zadanie 3

Zdefiniuj poniższą funkcję i sporządź jej wykres dla argumentów z danego przedziału:
$x>-1    oraz     x<1$
$x>-6    oraz     x<6$
$x>0     oraz     x<5$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def funkcja (przedzial):
    fun = lambda x: x**2 + 5
    x = np.linspace(przedzial[0], przedzial[1], 100)
    plt.figure()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Funkcja dla x w przedziale {}".format(przedzial))
    plt.plot(x, fun(x))
    legend = plt.legend(['Uzyskana funkcja'], title = "Legenda")
```

```python
przedzial1 = [-1, 1]
przedzial2 = [-6, 6]
przedzial3 = [0, 5]

funkcja(przedzial1)
funkcja(przedzial2)
funkcja(przedzial3)
```
